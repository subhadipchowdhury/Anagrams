```python
import collections
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
def anagram_fast(myword):
    return words_bysig[signature(myword)]
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
anagrams_all
```

```python
maxlen = max([len(word) for word in wordclean])
maxlen
```

```python
words_sublist_bysize = collections.defaultdict(dict)

for word in words_bysig.keys():
    words_sublist_bysize[len(word)].update({word : anagram_fast(word)})
    
nontrivial_anagrams_bysize = collections.defaultdict(dict)

for word,anagrams in words_bysig.items():
    if len(anagrams)>1:
        nontrivial_anagrams_bysize[len(word)].update({word : anagrams})
```
